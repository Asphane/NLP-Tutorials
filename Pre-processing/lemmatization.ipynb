```python
sent = "The cats were running faster and better than the dogs."
words = nltk.word_tokenize(sent)
pos_tags = nltk.pos_tag(words)

for word, tag in pos_tags:
    pos = get_pos(tag)
    print(word + " -> " + lem.lemmatize(word, pos))    

```
