```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

## Porter Stemmer

```python
from nltk.stem import PorterStemmer, SnowballStemmer, RegexpStemmer
ps = PorterStemmer()
```

```python
for word in words:
    print(word + "---->" + ps.stem(word))
```

## Snowball Stemmer

```python
sb = SnowballStemmer("english")
```

```python
for word in words:
    print(word + "---->" + sb.stem(word))
```

## RegexP Stemmer

```python
rs = RegexpStemmer(r'ing$|ed$|s$|ner$|s$')
words = ["playing", "played", "plays", "runner", "runs"]

for word in words:
    print(word + "---->" + rs.stem(word))
```
